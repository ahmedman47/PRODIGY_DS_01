Task: Create bar charts and histograms for categorical and continuous variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
df = pd.read_csv(url)
```

Inspect data

```python
print(df.head())
print(df.info())
```

Bar chart: Count of passengers by gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sex", data=df, palette="Set2")
plt.title("Count of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

Histogram: Distribution of passenger ages

```python
age_data = df["Age"].dropna()  

plt.figure(figsize=(8,5))
sns.histplot(age_data, bins=20, kde=True, color="skyblue")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

Distribution Plot

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df["Age"].dropna(), fill=True, color="purple", alpha=0.4)
plt.title("KDE Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```
